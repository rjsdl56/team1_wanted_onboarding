## (2) D_TYPE 확인

    - 원핫인코딩도 가능


- 생각정리
    - CC타입에만 가입일자 1970년이 있음 (전체 1093개)
    - 하지만 COUNT의 값은 그리 높지 않음 (가장 높은 것이 4 / 이상치로 보이는 값은 존재하지 않는다)
    - COUNT값이 낮고, 대부분의 값이 1인 것을 감안할 때 회사 직원보다는 시스템 오류일 확률이 큼
    - D_TYPE은 라벨인코더로 가야한다.
    - CC 타입일수록 COUNT의 값이 크다.
    - BB와 AA의 기준이 애매하다. (이건 팀원과 논의)
    - AA에는 압도적으로 많은 1의 값이 있지만, 이상치로 보이는 값의 절대적인 수는 BB보다 많다.
    - BB의 비율로 볼 때 이상치의 비율이 더 많다. 하지만 그 수가 매우 적다.

```python
# # 이제 각 AD1별 COUNT의 평균을 파악하자
# # 어차피 데이터 양으로 판단하는 것은 의미가 없음 ==> 1이 압도적
# # COUNT의 평균으로 보았을 때 높은 것들 위주로 선별하고, 이를 기준으로 binning?

check_dict = {}

for k in df['AD1'].unique() :
    new_df = df[df["AD1"]==k]
    check_dict[k] = new_df['COUNT'].mean()
print(check_dict)
```

```python
ad1_df = pd.DataFrame(check_dict, index = [0]) 
# index = [0] 안하면 ValueError: If using all scalar values, you must pass an index
ad1_df = ad1_df.T.reset_index().rename(columns={"index": "AD1"})
ad1_df.rename(columns={0:"mean"}, inplace=True)
display(ad1_df)
```
