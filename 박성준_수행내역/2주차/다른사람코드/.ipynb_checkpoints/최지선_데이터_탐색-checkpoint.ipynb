# 주제 : 앱 이용자별 향후 이용 건수 예측

### - Value chain상 정의
 - 이용자별 향후 이용 건수를 예측하여 마케팅, 영업 등의 의사결정 과정에 활용
 
### - 모델 개발 요건
 - 이용자별 결제 이력을 바탕으로 한 향후 한 달간 이용 건수 예측
 
### - Data 정의
> input : Tabular 데이터
 - 2020.01.01. ~ 2020.12.31. (결제일 기준 1년기간)
 
> output : Regression (향후 이용 건수 예측)

### - Task 정의
 - Regression, 유저별 한 달 후 이용 건수 예측
 
### - Metric 정의
 - MSE, MAE
 
### - 데이터 수집
 > - 유저 가입 정보 : ID, 가입일
 > - 소비 행동 정보 : 상품ID, 점포 주소, 결제일, 결제건수
 > - 외부 정보 : 서울시 지점별 교통량 데이터, 기온/강수량/풍속 등 기상데이터

## 모두의 주차장(모두컴퍼니) 비즈니스 모델 리서치
모두의 주차장’ 앱은 서울시 주차장정보 오픈API와 공공데이터포털에서 검색어 ‘주차장’을 입력했을 때 검색되는 30여 가지의 공공데이터를 모두 활용, 주차장정보를 체계적으로 정리해 DB화 시켰다. 이렇 게 DB화된 정보를 스마트폰을 통해 지도상에 표시, 주차장의 위치와 요금 정보를 쉽게 알아볼 수 있도 록 사용자에게 제공한다. 현재 위치를 중심으로 인근 주차장이 요금 정보와 함께 지도에 표시가 되기 때문에 위치와 요금을 직관적으로 비교할 수 있다. 타 서비스와 달리 사용자가 참여하여 직접 새로운 주차장정보를 제공하거나 오류에 대해 수정 요청도 할 수 있다. 또한 다양한 필터 기능으로 필요한 주 차장만 검색할 수 있고, 주차장 공유 기능을 제공하여 개인이 가지고 있는 주차 공간을 다른 사람들이 사용할 수 있도록 할 수 있다.

모두의 주차장은 공공데이터와 민간 주차장 정보를 바탕으로 주차장 정보 제공

주차장과 사용자를 연결해주는 커뮤니케이션 채널(공유서비스 플랫폼)

1) 거주자 우선 주차지역에 의존적인 수익



모두컴퍼니의 수익의 대부분은 공유주차장 중계 수수료를 통해 발생된다. 이 때 공유주차장은 거주자우선주차지역에 한정되어있다. 하지만, 거주자우선지역에 가장 활성화 되어있는 서울을 기준으로 약 16만 명이 운영되고 있다. (2015년 기준) 다시 말해서 모두의 주차장이 운영할 수 있는 최대 공유주차장은 거주자우선주차지역의 수에 한정된다는 뜻이다. 거주자우선주차지역은 주차공간의 소유권을 임시로 사용자가 일정기간동안 임대하는 형식을 가지기 때문에 지방자치단체 단위로 계약을 하게 되며, 수익 배분도 지자체와의 계약에 따라 다르다. 송파구 기준으로 약 16,000명의 거주자우선주차지역을 계약했을 때 수익이 약 1억 5천만원이다. 이를 기준으로 서울시의 거주자우선주차지역 수인 약 160,000만을 전부 관리한다 해도 약 15억원의 수익이 최대 수익으로 계산된다

### - DATE(결제일) 변수 활용하여 년,월,요일 새 변수 생성

```python
# DATE(결제일)을 년도와 월 나눠서 새 변수 생성
df_raw['DATE'] = pd.to_datetime(df_raw['DATE'])
df_raw['MONTH'] = df_raw['DATE'].dt.strftime('%m')
df_raw['YEAR'] = df_raw['DATE'].dt.strftime('%Y')
df_raw['WeekDay'] = df_raw['DATE'].dt.weekday # 0(월) ~ 6(일)
df_raw
```

```python
# 월별 결제건수

df_raw_c_month = df_raw['COUNT'].groupby(df_raw['MONTH']).sum()
df_raw_c_month = df_raw_c_month.reset_index()
df_raw_c_month = df_raw_c_month[['MONTH','COUNT']]
df_raw_c_month.columns = ['결제일(월)','결제건수']
df_raw_c_month
```

```python
plt.figure(figsize=(25,12))
color = sns.set_palette("Pastel2")
sns.barplot(df_raw_c_month['결제일(월)'], df_raw_c_month['결제건수'], palette=color)
plt.grid()
plt.title('결제일(월)별 결제건수(COUNT)', fontsize=25)
plt.ylabel('결제건수(COUNT)', fontsize=20)
plt.xlabel('결제일(월)', fontsize=20)
plt.show()
```

연초, 연말 겨울에 주차장 이용건수가 많을줄 알았는데 생각보다 1~2월의 이용건수가 적었다  
코로나 영향때문이었을까? 하반기로 갈수록 이용건수, 즉 결제건수가 많아지는걸 알 수 있다.

```python
# 요일별 결제건수

df_raw_c_wday = df_raw['COUNT'].groupby(df_raw['WeekDay']).sum()
df_raw_c_wday = df_raw_c_wday.reset_index()
df_raw_c_wday = df_raw_c_wday[['WeekDay','COUNT']]
df_raw_c_wday.columns = ['결제일(요일)','결제건수']
df_raw_c_wday
```

토,일의 주차장 이용건수가 많을 것 같았지만(외출 등으로) 일요일의 결제건수가 제일 낮음

#### 그렇다면 요일별 상품으로 추정되는 GOODS_TYPE과 위치로 추정되는 AD1의 결제건수를 살펴보자

```python
# 요일별 상품별 결제건수

df_raw_c_wday_G = df_raw.groupby(['WeekDay','GOODS_TYPE'])['COUNT'].sum().reset_index()
df_raw_c_wday_G = df_raw_c_wday_G[['WeekDay','GOODS_TYPE','COUNT']]
df_raw_c_wday_G.columns = ['결제일(요일)','상품','결제건수']
df_raw_c_wday_G
```

```python
# new_labels = {A:'당일권', B:'심야권', C:'휴일권', D:'종일권'}
# df_raw_c_wday_G['상품'] = df_raw_c_wday_G['상품'].map(new_labels)
fig, ax = plt.subplots(1,1, figsize = (15,8))
sns.pointplot(x = '결제일(요일)', y = '결제건수', hue = '상품', data = df_raw_c_wday_G, palette = color)
plt.title('요일별 상품별 결제건수', fontsize = 20, fontweight = 'bold')
plt.xlabel('결제일(요일)', fontsize = 12, fontweight = 'bold')
plt.ylabel('결제건수(건)', fontsize = 12, fontweight = 'bold')
plt.tick_params(axis = 'x', labelsize = 12)
plt.tick_params(axis = 'y', labelsize = 12)
fig.show()
```

전체적으로 상품 A의 결제건수가 가장 많았다.  
특히 상품A는 토요일에 결제가 많았고,  
상품B는 토요일과 일요일 주말에 결제건수가 현저히 낮아졌다

```python
# 요일별 지역별 결제건수

df_raw_c_wday_A = df_raw.groupby(['WeekDay','AD1'])['COUNT'].sum().reset_index()
df_raw_c_wday_A = df_raw_c_wday_A[['WeekDay','AD1','COUNT']]
df_raw_c_wday_A.columns = ['결제일(요일)','지역','결제건수']
df_raw_c_wday_A
```

```python
df_raw_c_wday_A.describe()
```

지역이 534개로 너무 많아서 결제건수 평균인 1694 이상 지역만 뽑아서 보자

```python
df_raw_c_wday_A_mean = df_raw_c_wday_A[df_raw_c_wday_A['결제건수']>1694]
df_raw_c_wday_A_mean
```

```python
# new_labels = {A:'당일권', B:'심야권', C:'휴일권', D:'종일권'}
# df_raw_c_wday_G['상품'] = df_raw_c_wday_G['상품'].map(new_labels)
fig, ax = plt.subplots(1,1, figsize = (15,8))
sns.pointplot(x = '결제일(요일)', y = '결제건수', hue = '지역', data = df_raw_c_wday_A_mean, palette = 'bright')
plt.title('요일별 지역별 결제건수', fontsize = 20, fontweight = 'bold')
plt.xlabel('결제일(요일)', fontsize = 12, fontweight = 'bold')
plt.ylabel('결제건수(건)', fontsize = 12, fontweight = 'bold')
plt.tick_params(axis = 'x', labelsize = 12)
plt.tick_params(axis = 'y', labelsize = 12)
fig.show()
```

가장 결제건수가 많은 지역 J로 추정되며, J같은 경우는 주말에 이용건수가 많이 낮아지는걸 알 수 있음  
반면 2번째로 결제건수가 많은 지역 GN은 주말에 J 지역보다 결제건수가 더 많다는 것을 알 수 있다.

AD1이 지역을 뜻하는지는 확실치 않으나, AD1이 지역이라며 꽤 중요한 변수로 보인다.
