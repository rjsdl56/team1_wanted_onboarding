```python
# 라이브러리 임포트
import os
import pandas as pd
import matplotlib.pyplot as plt
import sys

import warnings
warnings.filterwarnings('ignore')

print("파이썬 버전 :",sys.version)
print("판다스 버전 :",pd.__version__)
```

```python
# 데이터 확인
path = "C:/Users/parkgun/Desktop/프리온보딩/팀프로젝트_1주차/data"
print(os.listdir(path))

# 데이터 불러오기
df = pd.read_csv(path+"/company_nps_data.csv", encoding="cp949")
display(df.head())
```

```python
# 연도와 월을 묶어서 컬럼 형성
df['년월'] = pd.to_datetime(df['년도'].astype(str) + df['월'].astype(str), format='%Y%m').dt.strftime('%Y%m')

# 회사ID, 연도와 월을 제거
df.drop(df[['년도','월']], axis=1, inplace=True)

# 시간순으로 나열
df= df.sort_values(by='년월')
```

```python
display(df)
print(df.shape)
```

```python
# 코드 참고 : https://github.com/chsop/team4_wanted_onboarding/blob/main/Jaehwan_onboarding_week1_EDA_v0.1.ipynb
# 월별 연금 보험료와 월별 직원수가 0인 경우 제외 처리
df = df.drop(index=df.query('월별_직원수 == 0').index)
```

# 도전해보고자 하는 것.

- 인당 매출액을 계산
- 데이터프레임의 끝값만 모아서 딕셔너리 생성
- 값이 가장 큰 회사 10개 선정

```python
# 인당_연매출액 컬럼 생성
df['인당_연매출액'] = (df['연매출액'] / df['월별_직원수']).round(0)
display(df)
```

```python
# 아이디어 체크
new_dict = {}

df_391439 = df[df['회사ID']==391439]

# display(df_391439) # 이건 중간값이 가장 큰데?


# 일단 못먹어도 고
new_dict[391439] = df_391439.iloc[-1,-1]

print(new_dict)
```

```python
# 딕셔너리 생성 함수 만들기
def make_dict(df, x, new_dict) :
    new_df = df[df['회사ID']==x]
    new_dict[x] = new_df.iloc[-1,-1]
```

```python
# 회사별 마지막 매출액 추출

# for문용 list 생성
list_id = df['회사ID'].unique()

# 비어있는 딕셔너리 생성
sales_dict = {}

# 회사별 마지막 연매출액 설정
for k in list_id :
    make_dict(df,k,sales_dict)
```

```python
# 딕셔너리 값 높은 순으로 정렬
new_sales_dict = sorted(sales_dict.items(), key=lambda x: x[1], reverse=True)

# 가장 높은 10개 추출
top_10 = new_sales_dict[:10]

print(top_10)
```

```python
top_10[0][0]
```

```python
def visualization(df,x) :
    new_name = df[df['회사ID']==x]
    print(x,"회사 그래프입니다.")
    plt.figure(figsize=(20,10))
    plt.subplot(221)
    plt.plot(new_name['년월'],new_name['연매출액'])
    plt.title("매출액", fontsize=25)
    plt.xticks(rotation='vertical')
    plt.subplot(222)
    plt.plot(new_name['년월'],new_name['월별_직원수'])
    plt.title("직원수", fontsize=25)
    plt.xticks(rotation='vertical')
    plt.show()
    print()
    print()
```

```python
for z in range(len(top_10)) :
    check = top_10[z][0]
    
    visualization(df,check)
```
