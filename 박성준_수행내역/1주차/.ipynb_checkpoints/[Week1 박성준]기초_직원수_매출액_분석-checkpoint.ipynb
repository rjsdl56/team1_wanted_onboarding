```python
# 라이브러리 임포트
import os
import pandas as pd
import matplotlib.pyplot as plt
import sys

import warnings
warnings.filterwarnings('ignore')

print("파이썬 버전 :",sys.version)
print("판다스 버전 :",pd.__version__)
```

```python
# 데이터 확인
path = "C:/Users/parkgun/Desktop/프리온보딩/팀프로젝트_1주차/data"
os.listdir(path)
```

```python
# 데이터 불러오기
df = pd.read_csv(path+"/company_nps_data.csv", encoding="cp949")
display(df.head())
```

```python
# 시계열 데이터를 만들자

# 연도와 월을 묶어서 컬럼 형성
df['년월'] = pd.to_datetime(df['년도'].astype(str) + df['월'].astype(str), format='%Y%m').dt.strftime('%Y%m')

# 회사ID, 연도와 월을 제거해보자
df.drop(df[['년도','월']], axis=1, inplace=True)

# 시간순으로 나열해보자
df= df.sort_values(by='년월')

# 데이터 확인
display(df)
```

```python
# 직원수가 400명 넘어가는 기업
# 못 먹어도 고 ==> 생각보다 많네?

df_worker_400 = df[df['월별_직원수']>=400]
display(df_worker_400)
```

```python
print("직원수가 400명이 넘어가는 회사 목록 :",df_worker_400['회사ID'].unique())
print("직원수가 400명이 넘어가는 회사의 수 :",df_worker_400['회사ID'].nunique(), "개")
```

```python
# 데이터가 5개 이상인 회사만 남긴다.
new_df = df_worker_400.groupby("회사ID").filter(lambda x : len(x) > 5)

print("직원수가 400명이 넘어가는 회사 목록 :",new_df['회사ID'].unique())
print("직원수가 400명이 넘어가는 회사의 수 :",new_df['회사ID'].nunique(), "개")

# 회사ID 리스트
list_ID = new_df['회사ID'].unique()
```

```python
# 특정 회사만 빼서 만들어봅시다

set_name = 'df_ID_'
new_name = set_name + str(126814)
new_name = df[df['회사ID']==126814]

display(new_name)
```

```python
set_name = 'df_ID_'

x = 126814

new_name = set_name + str(x)

new_name = df[df['회사ID']==x]
print(x,"회사 그래프입니다.")
plt.figure(figsize=(20,10))
plt.subplot(221)
plt.plot(new_name['년월'],new_name['연매출액'])
plt.title("매출액")
plt.xticks(rotation='vertical')
plt.subplot(222)
plt.plot(new_name['년월'],new_name['월별_직원수'])
plt.title("직원수")
plt.xticks(rotation='vertical')
plt.show()
```

# 함수로 만들어서 회사매출액을 확인해봅시다.

```python
# 한번 직원수 기반 추출된 데이터로 매출액을 그려봅시다.

set_name = 'df_ID_'

for x in list_ID :
    new_name = set_name + str(x)
    
    new_name = df[df['회사ID']==x]
    print(x,"회사 그래프입니다.")
    plt.figure(figsize=(20,10))
    plt.subplot(221)
    plt.plot(new_name['년월'],new_name['연매출액'])
    plt.title("매출액")
    plt.xticks(rotation='vertical')
    plt.subplot(222)
    plt.plot(new_name['년월'],new_name['월별_직원수'])
    plt.title("직원수")
    plt.xticks(rotation='vertical')
    plt.show()
    print()
    print()
```
