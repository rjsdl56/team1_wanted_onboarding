# 찾아낸 인사이트

1. 데이터 내 모든 유저는 특정일에만 사용하고, 이후 재사용하지 않음 (order_date 값이 유일)
2. 자주 사용한 유저는 특정일에 다양하게 이용
3. brandnm 중 결측치는 sub_type 모두 TP
4. 제일 선호하는 브랜드는 구찌
5. p_type에 큰 차이는 없으니 4일권을 선호
6. 주로 이용하는 상품의 카테고리는 백
7. 1월에서 3월로 갈수록 많이 사용 (증가하는 추세)
8. 유저가 압도적으로 많음

# 분석목표

1. 구매 소요 시간 분석
    - 1-1. 가입일로부터 구매일까지 전환된 소요시간 분석
         - 최근 3개월 전체 분석 / 각 월 별 분석
         - 분기 경향 / 월 별 변동사항 확인 -> 트렌드 분석
             - ex. 가입 1일 이내 전환 유저 : 10% -> 15% 상승 성과
             
2. 구매 상품 트렌드 분석
    - 2-1. 유저당 주문 수량 분석
         - 최근 3개월 전체 분석 / 각 월 별 분석
         - 분기 경향 / 월 별 변동사항 확인 -> 트렌드 분석

    - 2-2. 주문 아이템 분석
         - 브랜드 별 선호도, 아이템 카테고리별 선호도 분석
         - 신규 유저들의 구매 트렌드 파악
         
         
3. 분석 내용 토대로 인사이트 도출
    - 3-1. 분석 내용 도식화
         - 1~2의 분석 내용 종합 및 도식화 (표 또는 그래프)
         - 유사성을 갖는 결과에 대하여 그룹화
    - 3-2. 인사이트 도출
         - 해당 결과를 통해 나타나는 사항 정리
         - 인사이트 도출 혹은 새로운 제안사항 전달 (없을 시 진행하지 않아도 무방)


용어 및 칼럼 설명
- brandnm : 아이템 브랜드명
- goodsnm : 아이템 명
- USER : 개별 유저 식별 값
- reg_date : 개별 유저 가입일
- order_date : 아이템 주문일
- p_type : 주문 상품 종류
     - 7days (7일권) / 4days (4일권)
- sub_type : 주문 상품 카테고리 
     - BA – 가방 / O – 아우터 / D – 드레스 / T – 상의 / B – 하의 / TB – 투피스 
- app_yn : 어플리케이션 사용 여부
     - Y – 어플리케이션 사용 / (공백) – 어플리케이션 미사용

```python
# 우선 p_type
sns.countplot(x ='p_type', data = df)
plt.title("이용권 전체 이용 추이")
plt.show()

# 전체로 볼땐 4일권을 더 많이 이용
df_p_type_7 = df[df['p_type']=='7days']
df_p_type_4 = df[df['p_type']=='4days']
print("7일권 이용자 비율 :", round(len(df_p_type_7)/len(df),2), "%")
print("4일권 이용자 비율 :", round(len(df_p_type_4)/len(df),2), "%")
```

```python
# sub_type
sns.countplot(x ='sub_type', data = df)
plt.title("상품 카테고리별 전체 이용 추이")
plt.show()

# 백 > 아우터 > 드레스

df_sub_type_BA = df[df["sub_type"]=="BA"]
df_sub_type_O = df[df["sub_type"]=="O"]
df_sub_type_B = df[df["sub_type"]=="B"]
df_sub_type_T = df[df["sub_type"]=="T"]
df_sub_type_D = df[df["sub_type"]=="D"]
df_sub_type_TP = df[df["sub_type"]=="TP"]

print("BA 이용자 비율 :", round(len(df_sub_type_BA)/len(df),2), "%")
print("O 이용자 비율 :", round(len(df_sub_type_O)/len(df),2), "%")
print("B 이용자 비율 :", round(len(df_sub_type_B)/len(df),2), "%")
print("T 이용자 비율 :", round(len(df_sub_type_T)/len(df),2), "%")
print("D 이용자 비율 :", round(len(df_sub_type_D)/len(df),2), "%")
print("TP 이용자 비율 :", round(len(df_sub_type_TP)/len(df),2), "%")
```
