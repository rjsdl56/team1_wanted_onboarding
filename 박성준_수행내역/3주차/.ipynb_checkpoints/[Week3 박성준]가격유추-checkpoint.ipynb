```python
# 결측치 보완
closet_share_df['brandnm'].fillna("Mixed", inplace=True)
closet_share_df['app_yn'].fillna("N", inplace=True)

# 결측치 제거
closet_share_df = closet_share_df.dropna(axis=0)
```

```python
# 의류데이터만 합산
df = closet_share_df[closet_share_df['sub_type']!="BA"]
```

```python
df.shape
```

```python
new_df = pd.merge(df, crawling_df, how="outer", on="goodsnm")
display(new_df.head())
```
